## 对比分析的定义

指两个或两个以上的对象在相同指标下进行对比分析，通过对比分析可以直观的看到变化和差距，从而找到业务不同阶段的问题。

【注意】绝对数据意义不大，要看相对数据

- 某产品的用户满意度达到了30%，这个结果是好还是不好？
- 单一的数值并不能说明什么问题，
- 因为并不能判断这个30%的满意度到底是高还是低？是升了还是降了？
- 只有通过对比分析，才能让数据变得有意义。

## 如何使用对比分析法
### 比什么？
- **绝对值**：本身具有价值的数据
  - 销售额
  - 阅读数
  - 视频/直播观看人数
- **比例值**：在具体环境中看比例才具有对比价值
  - 活跃占比：单独看月活跃数在上升，单看活跃占比却是下降的，说明用户粘性出现了问题。
  - 注册转化率


### 和谁比？
- 和自己比
  - 不同时期的自己比：从时间维度上对比现在和之前的数据差异，也称为趋势分析，用来追踪业务动态是否有异常
  - 同类比：同公司的不同业务线，比如教育公司的不同课程版块进行对比
  - 目标比：与既定的目标比，来了解当前数据距离目标的差距

- 和行业比
  - 和行业平均水平比，或者行业内标杆进行对比
  - 可清楚知道是自身因素还是行业趋势
    - 都涨，是否比同行涨的快
    - 都跌，是否比同行跌的慢

- 计划与实际比

### 怎么比？
- 数据整体大小
  - 常用平均组织、中位数
- 数据整体波动
  - 变异系数（标准差/平均值）
- 趋势变化
  - 常用时间折线图、同比、环比

## 注意事项
- 在使用对比分析时，如果对总体数值进行对比，那么**各对比维度的规模**需一致
  - 例如，A地区的销售额比其他地区差，我们不能简单的判断A地区店铺的销售能力不行，可能是因为A地区的店铺数量就比较少造成的




## 应用场景

#### 1. 新旧版本对比（产品改版评估）

判断产品上线、UI 改动、新功能是否有效。
如：新版注册流程 vs 旧版本 → 注册转化率对比。

---

#### 2. A/B 测试效果对比

对比两组实验的指标差异，帮助选择更优方案。
如：不同按钮颜色 → 点击率对比。

---

#### 3. 用户分群对比（找到关键行为）

比较不同用户群体的行为差异。
如：高留存用户 vs 低留存用户 → 找关键行为。

---

#### 4. 渠道效果对比（投放评估）

用于评估不同拉新渠道质量。
如：渠道 A 7 日留存 35% > 渠道 B 12%。

---

#### 5. 活动前后对比（活动有效性评估）

判断营销活动是否真正提升业务指标。
如：活动周 vs 平常周 → DAU/订单量变化。

---

#### 6. 时间周期对比（趋势判断）

用于判断业务是否增长，以及未来趋势
如：本周 vs 上周、本月 vs 上月、同比 vs 环比。

---

#### 7. 区域/品类对比（业务结构分析）

用于识别不同地区或品类表现差异。
如：华东 GMV > 华南 → 对应区域策略调整。

---

#### 8. 路径对比（转化与流失分析）

比较不同用户路径的表现，找高转化路径或流失节点。
如：

* 搜索 → 商品页 → 下单（转化 18%）
* 浏览 → 推荐 → 下单（转化 10%）

---

### 一句话总结

> **对比分析用于“找差异、评估效果、判断趋势”，是产品数据分析里最常用的方法之一，几乎所有评估类工作都离不开它。**
