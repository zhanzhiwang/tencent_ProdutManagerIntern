## AB test定义
简单来说，就是为同一个目标制定两个方案（比如两个页面），让一部分用户使用 A 方案，另一部分用户使用 B 方案，记录下用户的使用情况，看哪个方案更符合设计目标。当然，在实际操作过程之中还有许多需要注意的细节。

AB test可以找到因果关系。

#### AB test核心思想
1. 多个方案并行测试；

2. 每个方案只有一个变量不同；

3.  以某种规则优胜劣汰。


## AB test的引入
网站设计中，我们经常会面临多个设计方案的选择，比如某个按钮是用红色还是用蓝色，是放左边还是放右边。传统的解决方法通常是集体讨论表决，或者由某位专家或领导来拍板。虽然传统解决办法多数情况下也是有效的，但如果我们先试一试这样做的效果怎么样呢？A/B 测试（A/B Testing）可能是解决这类问题的一个更好的方法。

### AB test设计注意事项

- 需要特别留意的是第 2 点，它暗示了 A/B 测试的应用范围，——必须是单变量。

- 虽然 A/B 测试名字中只包含 A、B ，但并不是说它只能用于比较两个方案的好坏，事实上，你完全可以设计多个方案进行测试，比如ＡＢＣ测试，“A/B 测试”这个名字只是一个习惯的叫法。

- 以网站设计为例子

    - 一般来说，每个设计方案应该大体上是相同的，只是某一个地方有所不同，比如某处排版、文案、图片、颜色等。然后对不同的用户展示不同的方案。

    - 不同的用户在他的一次浏览过程中，看到的应该一直是同一个方案。比如他一开始看到的是 A 方案，则在此次会话中应该一直向他展示 A 方案，而不能一会儿让他看 A 方案，一会儿让他看 B 方案。

    - 还需要注意控制访问各个版本的人数，大多数情况下我们会希望将访问者平均分配到各个不同的版本上。要做到这些很简单，根据 cookie （比如 cookie 会话ID的最后一位数字）决定展示哪个版本就是一个不错的方法。

## AB test实施步骤
1. 确立目标

2. 设计实验

2. 开发两个或者多个版本

3. 收集数据

4. 分析数据，得出结果

## 分析数据的统计学方法

### 假设检验

做出两个或者多个版本进行测试时，不仅仅是版本不一样，而且版本面向的群体也是不一样的，即使群体可能是随机分配的很大程度上避免了因为群体划分造成的差异，但是差异一定是会有的，问题在于这种差异是否是偶然的，差异是否存在统计学上的意义呢？

这时我们需要用统计学上的假设检验来看这个问题。

 #### 假设检验步骤

 确定显著性水平-->确定原假设和备择假设-->确定统计量的分布-->计算p值-->得出结论

#### 例子1、单样本假设检验

某app用户使用时长是否高于行业水平？

根据调研报告，短视频app的行业平均日适用时长为4.9小时。

小A所在的公司一款短视频app口碑不错，想看看日用户使用时长是否高于行业水平。于是小A获取了公司100名用户的日使用时长数据（为什么不是所有用户，可能是搜集不到，这100个可能还是访谈得来的数据），这100名用户的平均值为5.5小时，标准差为2.87.


1. 确定显著性水平为0.05

2. 确定原假设和备择假设

原假设：没有高于行业水平

备择假设：高于行业水平

（这是一个单侧检验）

3. 收集数据

4. 数据分析，得出结论

根据中心极限定理总体分布的平均值为4.9，总体分布的标准差为$SD_{总体}=\frac{SD_{样本}}{\sqrt(样本量)} = \frac{2.87}{\sqrt{100}}=0.287$

即$\frac{\bar{X}-4.9}{SD_{样本} / \sqrt(样本量)} \sim t(100-1)$

$$ p = p(X > \frac{5.5-4.9}{SD_{样本}}) , 其中 X \sim t(99)$$

$$ p = p(X > \frac{5.5-4.9}{0.287}) =  p(X>2.09) = 0.0205$$

p值小于显著性水平0.05，所以拒绝原假设，表明小A公司的app日用户平均使用时长高于行业平均水平。

#### 例子2、双样本假设检验

共享单车公司希望进一步提高用户使用频率，运营的同时根据经验，猜测“发送单车使用券”可能会提升单车使用频率。因此数据分析师准备做一个AB test

1. 设定显著性水平

$\alpha = 0.05$

2. 确定原假设与备择假设

原假设：优惠券无提升

备择假设：优惠券有提升

3. 实验设计

同一区域、同一时间内，选取100名用户

随机分为两组：

实验组：发送优惠券

对照组：不发送优惠券

4. 收集数据

5. 数据分析

    - 原假设：$\bar{X}_T = \bar{X}_C$

    - 备择假设：$\bar{X}_T > \bar{X}_C$

    - 统计量：$\bar{X} = \bar{X}_T - \bar{X}_C$

    - 估计总体：
$$\mu_{\bar{X}} = 0$$
$$SD_{\bar{X}} \approx \sqrt{\frac{SD_1^2}{n_1} + \frac{SD_2^2}{n_2}} $$

$$\bar{X} \sim N(\mu_{\bar{X}}, SD_{\bar{X}})$$

   - 计算p值：
$$p = p(X > \frac{\bar{X}}{SD_{\bar{X}}}), X \sim N(\mu_{\bar{X}}, SD_{\bar{X}})$$
$$p=0.13$$

无法拒绝原假设，即发送单车使用券未必会显著提升用户的使用。

#### 倾向得分匹配
根据观察数据A和B，取得A和B中比较接近的两个子集A'和B'，模拟随机分配中的试验组和对照组，通过比较A'和B'，得出结论。

- 使用场景

两组混杂因素较多，可以是两组子集的取得更加合理。

- 如何取得两个接近的子集？

    - 罗辑回归模型$Y \sim X_1 + X_2 + \cdots$（Y为二分变量，1代表试验组，0代表控制组），$X_i$为混杂因素，会影响观测对象出现在A和B组的因素，这个案例中可能有性别、地区、身体健康状况等。

    - 对每个观测值，根据罗辑回归计算出罗辑回归值

    - 匹配A和B组中相似的的数据（罗辑回归值接近的个体），得到A'和B'，再比较两组的差异。

**注意**

临界罗辑回归值的选取应当控制在一个较窄的范围，这样选取的两组对象特征才会更加接近。

得到两组之后，在用T检验比较两组之间的差异是否具有统计意义上的显著性。

## AB test的应用场景


#### 1. 页面/UI 版本优化（最常见）

通过对比两个版本的界面，验证哪一个能提升核心指标。

* 不同按钮文案（如 “立即注册” vs “免费开始”）
* 不同按钮颜色/布局
* 不同 banner 展示方式
* 首页信息架构调整
* 登录/注册表单字段精简与否

**业务目的：提升转化率、完成率、点击率（CTR）。**

---

#### 2. 关键转化路径改版验证

在大幅改动核心流程前，通过 A/B 测对比改动收益。

* 下单流程优化（如减少步骤）
* 支付页面改动（展示优惠 or 不展示）
* 拉新流程（手机号注册 vs 第三方登录）
* 表单流程（减少必填项）

**业务目的：提升整体转化漏斗的关键“跳失点”。**

---

#### 3. Push/消息文案优化

* 推送标题 A vs B
* 推送时间 A vs B
* 推送场景触发方式对比

**业务目的：提升 Push 打开率、回流率、DAU。**

---

#### 4. 激励机制优化

* 发不发优惠券
* 优惠券金额 A/B
* 任务奖励金额差异
* 拉新活动奖励方式（现金 vs 积分）

**业务目的：促进转化、提高留存，但避免成本浪费。**

---

#### 5. 新功能灰度验证

功能上线前，先投放给部分用户进行测试：

* 首页新 tab
* 新的智能推荐模块
* 新的排序规则

**业务目的：验证新功能是否产生正向指标提升，降低风险。**

---

#### 6. 推荐排序/个性化算法评估

通过算法版本 A vs B 对比推荐效果。

* Feed 推荐排序策略
* Search 搜索结果排序方式
* 商品推荐策略（热销优先 vs 个性化优先）

**业务目的：提升 CTR、停留时长、GMV。**

---

#### 7. 定价策略测试

* 单品价格（如 A价 ¥19.9 vs B价 ¥29.9）
* 套餐/订阅价格
* 运费门槛（¥39 包邮 vs ¥49 包邮）

**业务目的：提升收益，但不损害转化率。**

---

#### 8. 留存类实验（用户运营）

* 新手引导方式不同
* 入职任务设计（有新手任务 vs 没有）
* onboarding 动画差异

**业务目的：提升 1 日留存、7 日留存。**

---

#### 9. 内容优化（内容型产品）

例如资讯、短视频、社区类：

* 不同标题形式（问句 vs 陈述句）
* 不同封面图/缩略图
* 不同推荐位布局

**业务目的：增加点击率、完读率、停留时长。**

---

#### 10. 广告位优化（广告变现类产品）

* 不同广告展示时机
* 插屏广告 vs 底部广告
* 广告数量控制

**业务目的：提升 eCPM 和广告收益，同时不会影响留存。**
