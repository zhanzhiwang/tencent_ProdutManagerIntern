## 漏斗分析法定义

漏斗模型是一套流程式数据分析，能够科学反映用户行为状态以及从起点到终点各阶段用户转化情况。

**漏斗分析的例子**

漏斗模型在商品购买中的应用，商品购买经过这些个过程 产看产品--使用产品--咨询价格--发送合同--购买产品。其中的每一个环节都有一定的转化率，我们需要做的事监控这些流程，并在其中寻找优化点，提高每一个流程的转化率。

### 建立漏斗时容易掉的坑

**1.漏斗一定要有时间窗口**
- 极端情况：23:59执行第一步，但次日00:01才执行第二步，那么这时这个步骤算完成还是没完成呢？以两日作为时间窗口就算做完成了。
- 如果我们不设置时间窗口，用户在2025.8.9执行了搜索商品，后来2025.12.5突然浏览并购买了商品，在步设置时间窗口的情况下，这算是完成了。这个时间跨度太长了，算是完成一个漏斗显然不对。
- 时间窗口的设置：根据具体业务来看
  - 按天：活动对用户的影响只在短期内有效（如断其活动）
  - 按周：业务本身复杂/决策成本高/多日才能完成（如理财/美股开户）
  - 按月：决策周期长（如装修买房）

时间窗口太长会包含很多无关信息，太短会扔掉很多有用信息

**2.漏斗一定是有严格顺序的**

如下，有一个漏斗流程是A-->B-->C-->D-->E

有人就会认为，反正都是看A到E的转化，那么A-->C-->E这个流程是否也可以用上面的流程观察？是不行的。

不可以用「ABCDE」漏斗，看「ACE」的数据。

因为A到C可能通过其他途径实现的，比如通过点击首页的推荐位、商品分类等，这样显然就与A-->B-->C途径不一样了。

**3.漏斗分析可以基于「用户」，也可以基于「事件」**

- 往往基于**用户**

- 也可以基于**事件**

**什么时候基于「用户」，什么时候基于「事件」呢？**

- 如果我是这门课程的负责人。我关心多少人选择我们并最终付费，那么我会基于用户分析。
- 如果我是这家企业的产品经理，负责很多门课程。我还会关心为什么有人第一次访问页面没有做出选择，而之后访问页面做出了选择呢？

## 漏斗分析场景

- 适用：有明确业务流程和业务目标
- 不适用：没有明确业务流程、跳转关系纷繁复杂的业务

---

### 一、产品侧应用场景

#### 1. 新用户注册/登录漏斗

用于查看新用户在哪一步流失，优化注册流程。

典型步骤：
曝光 → 点击注册 → 填写信息 → 验证 → 成功注册

应用场景：

* 找出哪个步骤转化最低
* 判断注册流程是否过长
* 优化引导、减少输入项



---

### 二、电商/交易类应用场景（高频）

#### 2. 交易漏斗分析（最经典）

典型步骤：
浏览商品 → 加入购物车 → 下单 → 支付成功

用于：

* 提升 GMV
* 找出下单或支付流失点
* 判断支付失败是否由技术或流程问题导致

---

#### 3. 营销活动效果评估（活动链路漏斗）

如 618、双 11、直播间活动等。

典型步骤：
活动曝光 → 点击进入 → 浏览活动页 → 下单 → 支付

用途：

* 评估活动 ROI
* 找出用户从 “兴趣” → “购买” 的瓶颈
* 优化活动页面、优惠策略

---

### 三、内容/社区产品应用场景

#### 4. 内容消费漏斗

用于分析内容平台的内容转化。

如：
曝光 → 点击 → 阅读 → 关注/点赞/评论

应用：

* 查看内容是否有吸引力
* 判断推荐算法质量
* 优化标题、封面、相关推荐

---

#### 5. 创作者转化漏斗（内容侧）

如小红书/抖音分析创作者路径：

发布内容 → 获得曝光 → 获得点击 → 获得互动 → 获得粉丝

用于：

* 评估平台给作者的流量倾斜效果
* 判断机制是否对新人友好

---

### 四、增长 & 用户运营场景

#### 6. 会员转化漏斗

用于会员营销策略优化。

例：
曝光会员权益 → 点击开通 → 查看权益说明 → 支付 → 成为会员

应用：

* 找出会员转化瓶颈
* 调整权益展示方式
* 识别付费敏感用户

---

#### 7. 用户复购漏斗

复购路径示例：
首购 → 浏览推荐 → 再次下单

应用：

* 找出复购驱动力
* 优化推荐策略、push 文案
* 提升长期投入产出（LTV）

---

#### 总结：一句话记住漏斗分析

> **漏斗分析用于找出用户流程中的转化瓶颈，通过逐步分析每个环节的流失，帮助提升注册、转化、留存、复购等核心业务指标。**
