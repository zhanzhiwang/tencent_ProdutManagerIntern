## 帕累托分析定义

帕累托分析法又称ABC分类法，也叫主次因素分析法。它是根据事物在技术和经济方面的主要特征，进行分类排队，分清重点和一般，从而对事物进行有区别地进行管理，提高效率。由于帕累托法则一般把对象划分为ABC三类，所以又称ABC分类法。

我们比较少听说帕累托分析，但是肯定听过二八法则，即20%的产品提供80%的收益。80%的商品只能提供20%的收益。而且在日常营销、管理中常常运用这种法则来把资源利用率得到最大化的实现。而二八法则则是基于帕累托法则的一个延申。

其表现形式为，在分析图表中，柱状图以显示各产品的销量或者库存等，一条曲线作为累计值呈现。对于左边纵坐标表示频数，右边纵坐标表示频率，以百分数表示。横坐标表示影响质量的各项因素，按影响大小从左向右排列，曲线表示各种影响因素大小的累计百分数。

这种方法有利于人们找出主次矛盾，有针对性地采取措施，提高效率减少成本。因为每个产品的频率频率或许有些差异，所以可以创建参数来动态地实现频率的调整。


## 帕累托分析步骤

1. 收集数据，针对不同的分析对象和分内容，收集有关数据

2. 统计汇总

3. 编制ABC分析表

4. ABC分析图

5. 确定重点管理方式

#### 例子分析

A类占了销售额的70%，这类商品在业务中至关重要，应当给予最高的管理优先级。

B类约占销售额的25%，贡献度也较高。

C类贡献率在5%左右，对整体业务影响较小，可见减少对这些商品的资源倾斜。

## 帕累托分析应用场景


#### 1. 用户投诉 / 客诉问题优先级排序（最常见）

将用户投诉类型按照数量排序，找出最主要的几个问题。

* 80% 的投诉可能来自 20% 的问题类型
  例如：
* 40% 来自“支付失败”
* 30% 来自“闪退”
* 15% 来自“收不到验证码”

→ 优先解决前两个即可显著减少投诉。

---

#### 2. 产品 Bug 的优先级管理

把所有 Bug 按影响用户数量排序：

* 影响大量用户的卡顿/崩溃/无法使用功能
* 影响少量用户但严重的 Bug
  → 先修那些能减少 80% 用户受影响的问题。

---

#### 3. 用户流失（Churn）原因分析

用户离开的原因往往不是平均分布的：

* 30% 因体验差
* 25% 因价格问题
* 20% 因找不到想要的内容

→ 针对前两个原因优化就能提升大部分留存。

---

#### 5. 内容型产品：头部内容贡献分析

内容平台（图文/短视频/资讯）中：

* 20% 的创作者贡献 80% 的流量
* 20% 的内容贡献 80% 的观看时长

→ PM 可以针对高价值创作者重点运营。

---

#### 6. 转化漏斗中丢单主要原因分析

例如电商：

* 下单失败的原因有很多
  但 80% 可能来自 1–2 个关键问题，例如：
* “库存不足”
* “填写地址繁琐”

→ 优先解决关键点，大幅提升 GMV。

---

#### 7. 推广渠道效果分析（渠道 80/20）

常见情况：

* 20% 的投放渠道带来 80% 的投放 ROI
* 20% 的渠道带来 80% 的新增用户

→ PM 可以更精准分配预算。

---

#### 8. 客户成功/企业服务：客户价值排序

对 C 端用户 OR B 端客户：

* 付费贡献、活跃度、使用深度并不平均
  帕累托分析可帮助：
* 识别 20% 的核心客户
* 加强重点服务、避免流失

---

#### 9. 需求池优先级排序

需求池中需求很多，但真正影响用户/业务的可能只有少部分。
通过帕累托图把需求按“业务价值”排序：

* 找出贡献 80% 价值的少数需求
  → 帮助 PM 做规划/ Roadmap 落地。

---

#### 总结

> “帕累托分析帮助产品经理在复杂的业务中快速定位最核心的 20% 问题，从而用有限资源产生最大的业务提升。它在投诉分析、Bug 优化、流失原因定位、增长渠道筛选、功能价值评估等场景中应用最广，能显著提升需求优先级判断和资源投入效率。”
